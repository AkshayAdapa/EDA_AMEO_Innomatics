## EXPLORATORY DATA ANALYSIS (EDA) & DATA VISUALIZATION ON AMEO DATASET

## EDA-->.

### EDA stands for Exploratory Data Analysis. It's an approach to analyzing datasets to summarize their main characteristics, often using statistical graphics and other data visualization methods. EDA aims to uncover underlying patterns, detect anomalies, and identify relationships between variables within the data. It's a crucial initial step in the data analysis process, helping analysts understand the structure and content of their datasets before applying more advanced techniques.

## Data Visualization

### Data visualization is the presentation of data in graphical or visual format. It transforms complex datasets into clear and concise visual representations such as charts, graphs, and maps. This visual approach helps to identify patterns, trends, and relationships within the data, making it easier for users to understand and interpret information quickly and effectively

### DATA DISCRIPTION:The dataset, released by Aspiring Minds from the Aspiring Mind Employment Outcome 2015 (AMEO), focuses on employment outcomes of engineering graduates. It includes variables like salary, job titles, and job locations, alongside standardized scores in cognitive, technical, and personality skills. With around 40 independent variables and 4000 data points, the dataset contains demographic features and academic details like graduation year, college tier, specialization, and GPA. Additionally, it features scores from sections of the AMCAT exam, covering English, logical reasoning, quantitative aptitude, and domain-specific skills. The dataset also includes scores from personality tests assessing conscientiousness, agreeableness, extraversion, neuroticism, and openness to experience.

## Univriant Analysis

#### Univariate analysis examines a single variable to understand its distribution and characteristics, providing insights into its central tendency, variability, and shape. Through visualizations and summary statistics, it reveals patterns and outliers within the data.

```python
# function for CDF
def cdf(data):
    x = np.sort(data)
    y = np.arange(1, len(x)+1)/len(x)
    return x, y
```

```python
def univariate_num(col):
    # Print basic statistics
    print(f"=== {col} Statistics ===")
    print(f"Median: {df[col].median()}")
    print(f"Maximum: {df[col].max()}")
    print(f"Minimum: {df[col].min()}")
    
    # Box plot
    plt.figure(figsize=(14, 10))
    plt.subplot(3, 2, 1)
    sns.boxplot(df[col])
    plt.title(f"Box Plot of {col}")
    
    # Histogram and KDE plot
    plt.subplot(3, 2, 2)
    plt.hist(df[col], color='blue', alpha=0.7, density=True)
    sns.kdeplot(df[col], color='green', label='KDE')
    plt.title(f"Histogram and KDE of {col}")
    plt.xlabel(col)
    plt.ylabel('Density')
    plt.legend()
    
    # CDF plot
    plt.subplot(3, 2, 3)
    x_tenure, y_tenure = cdf(df[col])
    x_sample_tenure, y_sample_tenure = cdf(np.random.normal(df[col].mean(), df[col].std(), size=len(df[col])))
    plt.plot(x_tenure, y_tenure, linestyle='-', marker='.', color='green', label=f"{col} CDF")
    plt.plot(x_sample_tenure, y_sample_tenure, linestyle='-', marker='.', color='red', label='Normal Distribution CDF')
    plt.title(f"CDF of {col}")
    plt.xlabel(col)
    plt.ylabel('CDF')
    plt.legend()
    
    # Violin plot
    plt.subplot(3, 2, 4)
    sns.violinplot(data=df, x=col, inner='quartile')
    plt.title(f"Violin Plot of {col}")
    plt.xlabel(col)
    
    plt.tight_layout()
    plt.show()
```
